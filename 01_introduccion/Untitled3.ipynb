```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.random.rand(1000, 1000, 3, 3)
```

```python
identity = np.identity(3, dtype=A.dtype)
```

```python
Ainv = np.zeros_like(A)
```

```python
Atrans = np.zeros_like(A)
```

```python
for i in range(1000):
    for j in range(1000):
        Ainv[i, j] = np.linalg.solve(A[i, j], identity)
        Atrans[i, j] = np.transpose(A[i, j])
```

```python
Atrans.shape
```

```python
Ainv.shape
```

```python
Atrans
```

```python
Ainv
```

```python

# suma de las matrices 
Ainv + Atrans
```

```python
# resta de las matrices 
Ainv - Atrans
```

```python

# Multiplicando  x  de las matrices 
Ainv @Atrans
```

```python
# Creando una matriz identidad de 2x2
I = np.eye(2) 
```

```python
# Multiplicar una matriz por la identidad nos da la misma matriz
A = np.array([[4, 7],
              [2, 6]])
```

```python
I
```

```python
A
```

```python
A @ I # AxI = A
```

```python
# Calculando el determinante de la matriz A
np.linalg.det(A)
```

```python

# Calculando la inversa de A.
A_inv = np.linalg.inv(A)
A_inv
```

```python
# A x A_inv nos da como resultado I.
A @ A_inv
```
